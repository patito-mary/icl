```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
from astropy.io import fits   # unambiguous import — avoids astropy.io namespace collision
from astropy.wcs import WCS

%matplotlib inline
```

```python
h5_file_halo = 'grid_halo_TNG100-1_99_111.hdf5'
fits_file_halo = 'grid_halo_TNG100-1_99_111.fits'
```

```python
with h5py.File(h5_file_halo, 'r+') as f:
    print("Keys: %s" % f.keys())
    # Get the data
    grid_data = f['grid'][...]
    print(grid_data)
    header_data = f['Header']
    #print(data)
```

```python
with h5py.File(h5_file_halo, 'r') as f:
    data = f['grid'][()]
    h5_hdr = dict(f['Header'].attrs)

# Build FITS header with HDF5 metadata
hdr = fits.Header()
hdr['SIM']     = (h5_hdr['sim_name'], 'IllustrisTNG simulation')
hdr['SNAP']    = (int(h5_hdr['snap']),  'Snapshot number')
hdr['HALO_ID'] = (int(h5_hdr['id']),   'Subhalo/halo ID')
hdr['TYPE']    = (h5_hdr['type'],      'Object type')
hdr['FILL_VAL']= (99.0,               'Sentinel value for empty pixels')
hdr['BUNIT']   = ('log SB',           'Pixel units (log surface brightness)')

hdu = fits.PrimaryHDU(data, header=hdr)
hdu.writeto(fits_file_halo, overwrite=True)
print(f"Written: {fits_file_halo}  shape={data.shape}  dtype={data.dtype}")
```
